# Linked List Cycle

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

### Technique: Two pointer technique!

Imagine you have a rabbit and a turtle, the rabit goes 2 steps faster than the turtle. If the list has a NUll value, then eventually the rabit + turtle will reach it and will stop. However, if there is no NULL, the rabbit and turtle will countinue to run until eventually they will meet at ***1 location. ***

### To solve this problem, create slow and fast variables
slow = head
fast = head.next

when slow == fast 
thats how both meet and the list has a cycle!


```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None 
class Solution: 
    def hasCycle(self, head: ListNode):
        
        #first chech if the head is empty - which means there is no cycle 
        if head ==None: 
            return False
        
        slow = head
        fast = head.next
        
        #IMPORTANT!!!! REMEMBER TO CHECK FAST AND SLOW if they are ever = to None then while loop returns False
        while(slow !=None and fast != None and fast.next !=None):
            # here we check the position to see if they match the values
            if fast==slow:
                return True
            #else we update the fast and slow to go to the next element
            fast = fast.next.next  # the fast will always go faster by 1 or more until it catches up to slow
            slow= slow.next
        return False
```
