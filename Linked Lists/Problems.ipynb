```python
#add a node at the beginning 
class Node:
    def __init__(self, val=None, next = None):
        self.val = val
        self.next = next 
    
class LinkedList: 
    def __init__(self):
        self.head = None
        
    #add a value to the begining 
    def prepend(self, val):
    # intitialize a new node cur
    # Link the new node to our riginal head node head
    # Assign cur to head 
        newNode = Node(val)
        self.head = newNode
        
    #add a value at an index 
    def append(self, val):
        #initialize a new node with the given value
        # link the next field of node to the prev;s new node 'next'
        #link the next field in prev to cur 
        newNode= Node(val)
        prev = self.head
        prev.next = newNode.next
        
    def printLL(self):
        if self.head is None: 
            print('Linked list is empty')
            return 
        iteration = self.head 
        list_string = ''
        while iteration:
            list_string +=str(iteration.val)+'-->'
            iteration = iteration.next #link the head to the next node
        print(list_string)
        
if __name__=='__main__':
    ll = LinkedList()
    ll.prepend('C')
    ll.append('B')
    ll.printLL()
```
